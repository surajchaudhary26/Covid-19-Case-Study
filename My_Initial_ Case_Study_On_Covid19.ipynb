**Let's prepair data for the model**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
daily_cases = df.groupby('Date')['Confirmed'].sum().reset_index()
```

```python
daily_cases['Date'] = pd.to_datetime(daily_cases['Date'])
daily_cases['days_since_start'] = (daily_cases['Date'] - daily_cases['Date'].min()).dt.days
```

```python
# First, we define our input (X) and our output (y)
X = daily_cases[['days_since_start']]
y = daily_cases['Confirmed']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Plot the actual data
plt.figure(figsize=(12, 6))
```

```python
# 1. Plot the training data (the part the model learned from)
plt.plot(X_train['days_since_start'], y_train, label='Training Data (Actual)')

# 2. Plot the testing data (the part we're testing on)
plt.plot(X_test['days_since_start'], y_test, color='orange', label='Test Data (Actual)')

# 3. Plot the model's predictions
plt.plot(X_test['days_since_start'], y_pred, color='red', linestyle='--', linewidth=2, label='Model Predictions')



plt.title('COVID-19 Confirmed Cases: Actual vs. Predicted')
plt.xlabel('Days Since Start')
plt.ylabel('Total Confirmed Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
